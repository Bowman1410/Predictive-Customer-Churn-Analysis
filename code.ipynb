Visualization To understand the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Telco-Customer-Churn-Data.csv"  # Replace with the actual path to your CSV file
df = pd.read_csv(file_path)

# Convert 'TotalCharges' to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Handle missing values if needed
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)

# Convert 'Churn' column to numerical values (if not already)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Select relevant columns for correlation analysis
correlation_data = df[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']]

# Check for missing values
print(correlation_data.isnull().sum())

# Handle missing values if needed
correlation_data.fillna(correlation_data.mean(), inplace=True)

# Check for non-numeric values
print(correlation_data.dtypes)

# Convert to numeric
correlation_data = correlation_data.apply(pd.to_numeric, errors='coerce')

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix - Factors Correlated with Churn')
plt.show()

# Demographic Analysis

# Bar chart for Gender distribution
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Count plot for Senior Citizens
plt.figure(figsize=(8, 6))
sns.countplot(x='SeniorCitizen', hue='Churn', data=df, palette='pastel')
plt.title('Senior Citizen Distribution')
plt.xlabel('Senior Citizen')
plt.ylabel('Count')
plt.show()

# Stacked bar chart for Partner and Dependents
partner_dependents_counts = df.groupby(['Partner', 'Dependents']).size().unstack()
partner_dependents_counts.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'], figsize=(10, 8))
plt.title('Distribution of Partners and Dependents')
plt.xlabel('Partner')
plt.ylabel('Count')
plt.legend(title='Dependents', loc='upper right')
plt.show()

def churn_analysis_bar_chart(feature):
    churn_counts = df.groupby([feature, 'Churn']).size().unstack()
    churn_counts.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'])
    plt.title(f'Churn Analysis by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.legend(title='Churn', labels=['No', 'Yes'])
    plt.show()

# Example usage for gender
churn_analysis_bar_chart('gender')
```

Pie Chart Showing the Distribution of Monthly Charges and Total Charges

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the dataset in a CSV file named 'Telco-Customer-Churn-Data.csv'
# Adjust the file path accordingly
file_path = 'Telco-Customer-Churn-Data.csv'

# Load the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)

# Convert 'MonthlyCharges' and 'TotalCharges' columns to float
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Drop rows with missing values in 'MonthlyCharges' or 'TotalCharges'
df = df.dropna(subset=['MonthlyCharges', 'TotalCharges'])

# Extract the 'MonthlyCharges' and 'TotalCharges' columns
monthly_charges = df['MonthlyCharges']
total_charges = df['TotalCharges']

# Create a pie chart
labels = ['Monthly Charges', 'Total Charges']
sizes = [monthly_charges.sum(), total_charges.sum()]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # explode the 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

# Display the pie chart
plt.title('Distribution of Monthly Charges and Total Charges')
plt.show()
```

The Scatter Plot  represents the relationship between 'Monthly Charges' and 'Total Charges' for each customer in the dataset

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your dataset is loaded into a Pandas DataFrame named 'df'
# You can load your dataset with the following code:
df = pd.read_csv('Telco-Customer-Churn-Data.csv')

# Extract 'MonthlyCharges' and 'TotalCharges' columns
monthly_charges = df['MonthlyCharges']
total_charges = pd.to_numeric(df['TotalCharges'], errors='coerce')  # Convert 'TotalCharges' to numeric, handle errors as 'coerce'

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(monthly_charges, total_charges, alpha=0.5)
plt.title('Relationship between Monthly Charges and Total Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.grid(True)
plt.show()


```
